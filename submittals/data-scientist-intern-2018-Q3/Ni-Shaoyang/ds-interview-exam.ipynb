### Q 1.1: Fibionacci
![fib image](https://upload.wikimedia.org/wikipedia/commons/thumb/9/93/Fibonacci_spiral_34.svg/200px-Fibonacci_spiral_34.svg.png)

#### Q 1.1.1
Given $n$ where $n \in \mathbb{N}$ (i.e., $n$ is an integer and $n > 0$), write a function `fibonacci(n)` that computes the Fibonacci number $F_n$, where $F_n$ is defined by the recurrence relation:

$$ F_n = F_{n-1} + F_{n-2}$$

with initial conditions of:

$$ F_1 = 1,  F_2 = 1$$

```python
def fibonacci(n): # Recursive Method
    if n == 1 or n == 2:
        return 1 
    else:
        return fibonacci(n-1) + fibonacci(n-2)
```

#### Q 1.1.2
What's the complexity of your implementation?

The complexity of the your implementation is :\begin{equation*} O(2^n) \end{equation*}, so the running go 

```python
import time
import matplotlib.pyplot as plt
running_time = []
for i in [5,10,20,40]:
    start_time = time.time()
    fibonacci(i)
    running_time.append(time.time() - start_time)
plt.plot([5,10,20,40],running_time)
plt.title("Request Number N vs Running Time for Fibonacci")
plt.xlabel("Request Number N")
plt.ylabel("Running Time")
plt.show()
```

#### Q 1.1.3
Consider an alternative implementation to compute Fibonacci number $F_n$ and write a new function, `fibonacci2(n)`.

```python
def fibonacci2(n):
    fn1 = 1
    fn2 = 1
    while n != 2:
        fn1, fn2 = fn1+fn2, fn1
        n-=1
    return fn1
```

#### Q 1.1.4
What's the complexity of your implementation?

The complexity of the your implementation is :\begin{equation*} O(n) \end{equation*},so the running time grows linearly as the number of request time n. 

```python
running_time_2 = []
for i in [20,40,60,80,100]:
    start_time = time.time()
    fibonacci2(i)
    running_time_2.append(time.time() - start_time)
plt.plot([20,40,60,80,100],running_time_2)
plt.title("Request Number N vs Running Time for Fibonacci2")
plt.show()
```

#### Q 1.2.1
Consider a [singly linked list](https://en.wikipedia.org/wiki/Linked_list), $L$. Write a function `is_palindrome(L)` that detects if $L$ is a [palindrome](https://en.wikipedia.org/wiki/Palindrome), by returning a bool, `True` or `False`.


```python
def is_palindrome(L):
    slow = L 
    fast = L.next
    val_slow.append(slow.val)
    while slow and fast:
        slow = slow.next
        fast = fast.next.next
        val_slow.append(slow.val)
        
    if fast is None:
            val_slow.pop()
            
    while slow.next:
        if slow.next.val != val_slow.pop():
            return False
        slow = slow.next
        
    return True   
```

#### Q 1.2.2
What is the complexity of your implementation?

The complexity of the function is_palindrome is $$O(n)$$ since the total number of loops in the while loop is n. 

#### Q 1.2.3
Consider an alternative implementation to detect if L is a palindrome and write a new function, `is_palindrome2(L)`.

```python
def is_palindrome2(L):
        lst = []
        while L != None:
            lst.append(L.val)
            L = L.next
        
        left, right = 0, len(lst) -1
        
        while left < right:
            if lst[left] != lst[right]:
                return False
            left += 1
            right -= 1
        return True
```

#### Q 1.2.4
What's the complexity of this implementation?


The complexity of the function is_palindrome is: $$O(n)$$, which is the same as is_palindrome(). The difference between first and second function is that first method check if the linked list is palindrome from the middle, and the second method check if the linked list is palindrome from the both sides of the list.

### Q 2.1: Finding $\pi$ in a random uniform?
![pi pie img](http://core2.staticworld.net/images/article/2016/03/pi-day-intro-100649273-carousel.idge.jpeg)

Given a uniform random generator $[0,1)$ (e.g., use your language's standard libary to generate random value), write a a function `compute_pi` to compute [$\pi$](https://en.wikipedia.org/wiki/Pi).

```python
import numpy as np

def compute_pi(n):
    x = np.random.uniform(0,1,n)
    y = np.random.uniform(0,1,n)
    return 4 * sum((x**2 + y**2)**0.5 <= 1) / n
n = 10000
print("The estimated pi after {:d} simulations is {:0.3f}:".format(n,compute_pi(n)))
```

### Q 2.2: Making a 6-side die roll a 7?

![reno die image](http://thumbs.ebaystatic.com/images/g/IQ8AAOSwvzRXyagD/s-l225.jpg)

Using a single 6-side die, how can you generate a random number between 1 - 7?

```python
def roll_dice_get_7():
    dice_num = np.random.randint(1,7)
    gen_num_range = [dice_num,dice_num+1]

    determine_num = np.random.randint(1,7)
    if determine_num >= 4:
        return gen_num_range[1]
    else:
        return gen_num_range[0]
```

### Q 2.3: Is normality uniform?

![normal and uniform distributions](https://qph.ec.quoracdn.net/main-qimg-f6ed71ed1d0059760fb63db384dcbcca-c)

Given draws from a normal distribution with known parameters, how can you simulate draws from a uniform distribution?

The methond I use to transform repairs of draws (X,Y) from a normal distribution is by calculating the angle represented by (X,Y) in the circle. As we know, any pair of draws (X,Y) from normal distribution will drop in a circle with certain radius. 

```python
import numpy as np
x = np.random.normal(0, 1, 10000)
y = np.random.normal(0, 1, 10000)
plt.scatter(x=x,y=y)
plt.xlim(-5, 5)
plt.ylim(-5, 5)
plt.gca().set_aspect('equal', adjustable='box')
plt.show()
```

In addition, the probability of that any pair of draw drops anywhere in a circle is equal. Therefore, we use this fact to generate a uniform distribution.

```python
r = (x**2 + y **2)**0.5 
deg = np.arcsin(y/r)*2/np.pi
num_each_group = []
for i in range(0,4):
    num_each_group.append(sum((deg >= (-1 + 0.5 * i)) & (deg< (-1 + 0.5 * (i+1)))))

plt.bar(np.arange(len(num_each_group)),num_each_group,align='center', alpha=0.5)
plt.title("Simulate Uniform Distribution from a Normal Distribution")
plt.show()
```

### Q 2.5: Uber vs. Lyft

![uber vs lyft](http://usiaffinity.typepad.com/.a/6a01347fc1cb08970c01bb0876bcbe970d-pi)

You request 2 UberXâ€™s and 3 Lyfts. If the time that each takes to reach you is IID, what is the probability that all the Lyfts arrive first? What is the probability that all the UberXâ€™s arrive first?

Since the time that each car takes to reach you is IID, so the conditional probability which car will come first is the same. In order to clarify the solution, let's label that:<br>
    1. A: the first three cars comes are all Lyft
    2. A1: the first car comes is Lyft
    3. A2: the second car comes is Lyft when the first car comes is Lyft
    4. A3: the third car comes is Lyft when the first two cars come is Lyft
so the probability of that all Lyfts arrive first is:
    $$ P(A) = P(A1) * P(A2) * P(A3) = (\frac{3}{5}) * (\frac{2}{4}) * (\frac{1}{3}) = \frac{1}{10}$$   
The same method applies to the case that all the UberX arrive first. The probability can be calculated as: 
    $$ P(B) = P(B1) * P(B2) = (\frac{2}{5}) * (\frac{1}{4}) = {\frac{1}{10}}$$
